```python
buro = pd.read_csv('../input/bureau.csv')
buro = buro[['SK_ID_CURR','SK_ID_BUREAU']]
buro.head()
```

```python
buro_balance.loc[buro_balance['STATUS']=='C', 'STATUS'] = '0'
buro_balance.loc[buro_balance['STATUS']=='X', 'STATUS'] = '0'
buro_balance['STATUS'] = buro_balance['STATUS'].astype('int64')
```

```python
buro_balance.head()
```

```python
buro_balance_group = buro_balance.groupby('SK_ID_BUREAU').agg({'STATUS':['max','mean','min','sum'],'MONTHS_BALANCE':['max','count']})
buro_balance_group.columns = [' '.join(col).strip() for col in buro_balance_group.columns.values]
buro_balance_group
```

```python
buro_balance_group.columns
```

```python
buro_balance_group[buro_balance_group['STATUS max']==5]
```

```python
buroMerged = buro.merge(buro_balance_group, left_on='SK_ID_BUREAU',right_index=True, how='left', suffixes=['', '_BALANCE_HIST'])
buroMerged.sort_values('SK_ID_BUREAU')
```
