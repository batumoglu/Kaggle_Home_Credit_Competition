```python
def generateDistribution(df):
    for i in range(len(df.columns)):
        if(df.iloc[:,i].dtypes=='object'):
            print(df.columns[i]+ ' is object')
            continue
        try:
            plt.figure(figsize=(12,8))
            type(df.iloc[:,i])
            sns.distplot(df.iloc[:,i], bins=50, kde=False)
            plt.show()
        except:
            print(df.columns[i])
```
